## **1. Emotion Classification Model**

```python
import os
#os.chdir("/content")
cwd = os.getcwd()
os.chdir(cwd)
```

```python
emotions_path = cwd + "/Emotions"
os.chdir(emotions_path)
```

```python
# Imports
from model import BertForMultiLabelClassification
import torch

from attrdict import AttrDict
import json
from transformers import BertTokenizer
from multilabel_pipeline import MultiLabelPipeline
from pprint import pprint

from evaluate_model import evaluate_model
```

Load pre-trained model

```python
path = os.getcwd() + "/checkpoint-27000"
os.chdir(path)
checkpoint = os.path.dirname(path + "/pytorch_model.bin")

model = BertForMultiLabelClassification.from_pretrained(checkpoint)

os.chdir(emotions_path)
config_filename = "original.json"
with open(os.path.join("config", config_filename)) as f:
    args = AttrDict(json.load(f))
args.output_dir = os.path.join(args.ckpt_dir, args.output_dir)
```

Evaluate pre-trained model on test dataset

```python
# Download model evaluation file
evaluate_model(config_filename, checkpoint)
```

Test pre-trained model on unseen data

```python
tokenizer_path = os.path.dirname(path + "/tokenizer_config.json")
tokenizer = BertTokenizer.from_pretrained(tokenizer_path)

goemotions = MultiLabelPipeline(
    model=model,
    tokenizer=tokenizer,
    threshold=0.3
)

texts = [
    "Hey that's a thought! Maybe we need [NAME] to be the celebrity vaccine endorsement!",
    "it’s happened before?! love my hometown of beautiful new ken 😂😂",
    "I love you, brother.",
    "Troll, bro. Th    ---ey know they're saying stupid shit. The motherfucker does nothing but stink up libertarian subs talking shit",
]
# This will output the labels and corresponding score found in each of the sentences above
pprint(goemotions(texts))
```
